## Import required Python libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read Wine Quality dataset
data = pd.read_csv(r"..\Data\wine_quality.csv", index_col=0)
data.head(5)
data.shape
```

```python
# Statistical description of the dataset
data.describe()
```

# Generate target labels for classification

The wine quality attribute in the dataset is a score in a scale of 10. For classifying the wine we are going to label the examples with score higher than 7 as good quality.

```python
data['grade'] = 1 # High quality
data.grade[data.quality < 7] = 0 # Poor quality

plt.figure(figsize = (8,8))
labels = data.grade.value_counts().index
plt.pie(data.grade.value_counts(), autopct='%1.1f%%')
plt.legend(labels, loc="best")
plt.axis('equal')
plt.title('Quality Pie Chart')
plt.show()
```

# Classifying wine quality 

## Import scikit-learn classifiers

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
import xgboost
from sklearn.metrics import accuracy_score
```

## Prepare data for classifier

```python
data_features = data.drop(['quality', 'grade', 'color'], axis=1) # Input features matrix

# Split data into training and test set
x_train, x_test, y_train, y_test = train_test_split(data_features, data['grade'], test_size=0.1, random_state=7)
```

```python
x_train_values = x_train.values
x_test_values = x_test.values
```

## Implement your own wine classifier! 

```python
# Implementation


```

## Example: Decision Tree classifier

```python
decisiontree = DecisionTreeClassifier(max_depth=4)
decisiontree.fit(x_train_values, y_train)
decisiontree_pred = decisiontree.predict(x_test_values)
```

```python
plt.figure(figsize = (20,8))
domain = np.linspace(1, 100, len(decisiontree_pred)) 
plt.plot(domain, decisiontree_pred,'o')
plt.plot(domain, y_test,'o')
plt.legend(('Prediction','Actual value'))
plt.show()
```

## Accuracy of the Decision Tree

```python
print('Accuracy:', accuracy_score(y_test, decisiontree_pred)*100,'%')
```

## Finding the most relevant features to predict wine quality

```python
importances = decisiontree.feature_importances_
indices = np.argsort(importances)
```

```python
plt.figure(1)
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), data_features.columns[indices])
plt.xlabel('Relative Importance')
```
